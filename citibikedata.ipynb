```python
import pandas as pd
import datetime
```

```python
july_seventeen = pd.read_csv("Data/201707-citibike-tripdata.csv")
july_seventeen
```

```python
july_eighteen = pd.read_csv("Data/201807-citibike-tripdata.csv")
july_eighteen
```

```python
july_nineteen = pd.read_csv("Data/201907-citibike-tripdata.csv")
july_nineteen
```

```python
july_twenty = pd.read_csv("Data/202007-citibike-tripdata.csv")
july_twenty
```

```python
combined_data = pd.concat([july_seventeen, july_eighteen, july_nineteen, july_twenty], ignore_index=False)
combined_data
```

```python
combined_data ['Date'] = pd.to_datetime(combined_data['starttime']).dt.date
combined_data ['Time'] = pd.to_datetime(combined_data['starttime']).dt.time
combined_data
```

```python
combined_data.dropna()
combined_data
```

```python
combined_data.to_csv("Data/CitiBike_Data.csv")
```
